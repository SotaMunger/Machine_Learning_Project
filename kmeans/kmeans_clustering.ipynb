## Transaction Data

```python
file_path = "C:/Users/conor/OneDrive/Documents/Python Scripts/Machine Learning/Project/Translated Data/Translated Data/transactions(SU_ORNEKLEM_KK_HAR_BILGI).csv"
transaction_df = pd.read_csv(file_path, dtype={'Category': 'str'}, encoding = "utf-8")
transaction_df
```

# KMeans Clustering

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline([
('imputer', SimpleImputer(strategy="median")),
('std_scaler', StandardScaler())
])
```

```python

from sklearn.preprocessing import OneHotEncoder

cat_pipeline  = Pipeline([
('cat_encoder', OneHotEncoder())
])
```

```python
from sklearn.compose import ColumnTransformer

cat_attribs = [ 'Gender', 'Education_Lv', 'Job_Status', 'Marital_Status']
num_attribs = [x for x in full_df.columns if x not in cat_attribs]

full_pipeline = ColumnTransformer([
("num", num_pipeline, num_attribs),
("cat", cat_pipeline, cat_attribs), 
], remainder = 'passthrough')
```

```python
scaled_full_df = pd.DataFrame(full_pipeline.fit_transform(full_df))
scaled_full_df
```

```python
from sklearn.cluster import KMeans

k = 5
X = scaled_full_df.to_numpy()
kmeans = KMeans(n_clusters= k, random_state = 16)
kmeans.fit_predict(scaled_full_df)
```

```python
kmeans_per_k = [KMeans(n_clusters=k, random_state = 16).fit(scaled_full_df)
                for k in range(1, 11)]
inertias = [model.inertia_ for model in kmeans_per_k]
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

plt.figure(figsize=(8, 3.5))
plt.plot(range(1, 11), inertias, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Inertia", fontsize=14)
plt.axis([1, 10.5, 3000000, 5000000])
plt.savefig('full_inertias.png')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_scores = [silhouette_score(scaled_full_df, model.labels_)
                     for model in kmeans_per_k[1:]]
```

```python
plt.figure(figsize=(8, 3))
plt.plot(range(1,10), silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette score", fontsize=14)
plt.axis([1, 10.5, 0, 0.75])
plt.savefig('full_scores.png')
plt.show()
```

```python
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter

plt.figure(figsize=(11, 9))

for k in (2, 3, 4, 5):
    plt.subplot(2, 2, k-1)
    
    y_pred = kmeans_per_k[k - 1].labels_
    silhouette_coefficients = silhouette_samples(scaled_full_df, y_pred)

    padding = len(X) // 30
    pos = padding
    ticks = []
    for i in range(k):
        coeffs = silhouette_coefficients[y_pred == i]
        coeffs.sort()

        color = mpl.cm.Spectral(i / k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs) // 2)
        pos += len(coeffs) + padding

    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    if k in (4, 6):
        plt.ylabel("Cluster")
    
    if k in (6, 7):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel("Silhouette Coefficient")
    else:
        plt.tick_params(labelbottom=False)

    plt.axvline(x=silhouette_scores[k - 2], color="red", linestyle="--")
    plt.title("$k={}$".format(k), fontsize=16)
plt.savefig('full_diagrams.png')
plt.show()
```
